```python
# Install necessary libraries
!pip install transformers

from transformers import pipeline
import pandas as pd
```

```python
# Sample data (Replace this with your dataset)
data = [
    "I love using machine learning for sentiment analysis!",
    "This project is quite challenging but rewarding.",
    "I hate encountering errors in my code.",
    "The tutorial was very helpful and easy to follow."
]
```

```python
# Load BERT-based sentiment analysis pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
# Perform sentiment analysis
results = sentiment_pipeline(data)
```

```python
# Display results
sentiment_df = pd.DataFrame(results)
print(sentiment_df)
```

```python
# Save results to CSV
sentiment_df.to_csv("sentiment_analysis_results.csv", index=False)
```
